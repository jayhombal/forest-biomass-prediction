## Scaling and Train Test Split

```python
X = matched_trees.drop('agb',axis=1)
y = matched_trees['agb']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
lrmodel = LinearRegression()
results = lrmodel.fit(X_train_scaled, y_train)
```

```python
results.coef_
```

```python
y_predict = lrmodel.predict(X_test_scaled)
```

```python
print(lrmodel.intercept_, lrmodel.coef_)
print(math.sqrt(mean_squared_error(y_test, y_predict)))
print((r2_score(y_test,y_predict)))
```

```python
print(lrmodel.score(X_test, y_test))
```
